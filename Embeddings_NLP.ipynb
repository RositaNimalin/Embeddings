```python
Text = '''Natural Language Processing (NLP) is a fascinating field that combines computer science and linguistics to enable machines\
 to understand, interpret, and generate human-like text. NLP applications are diverse, ranging from sentiment analysis and machine translation \
 to chatbots and text summarization. As the demand for intelligent language processing systems grows, the importance of NLP embeddings becomes \
 evident.Embeddings are numerical representations of words or phrases in a continuous vector space, capturing semantic relationships between them.\
 Word embeddings, such as Word2Vec, GloVe, and FastText, have become essential tools for various NLP tasks. These embeddings transform words into \
 vectors, allowing algorithms to work with semantic information rather than mere syntactic patterns.In this era of deep learning, pre-trained \
 language models like BERT, GPT, and RoBERTa have gained prominence. These models generate contextualized embeddings that capture the meaning of \
 words based on their context within a sentence. Leveraging pre-trained embeddings can significantly boost the performance of NLP models, as they \
 learn from vast amounts of textual data.In your NLP embeddings code,consider using popular libraries like spaCy, gensim, or TensorFlow to load and \
 manipulate embeddings. You can experiment with different embedding models and fine-tune them for your specific task. Additionally, visualization \
 techniques, such as t-SNE, can help you explore the embedding space and gain insights into the relationships between words.Whether you're working \
 on text classification, named entity recognition, or any other NLP task, choosing the right embeddings is crucial for achieving optimal performance. \
 As the field of NLP continues to advance, staying updated on the latest embedding techniques and incorporating them into your code will \
 contribute to the success of your language processing projects.'''
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
sent_token = sent_tokenize(Text)
```

```python
sent_token
```

# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer()
X = bow.fit_transform(sent_token).toarray()
print(bow.get_feature_names_out())
print(X[:10])
```

```python
X[0].shape
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(sent_token).toarray()
print(tfidf.get_feature_names_out())
```

```python
X[:10]
```

# Word2Vec

```python
import gensim
from gensim.models import Word2Vec
```

```python
model1 = gensim.models.Word2Vec([word_tokenize(sent_token[0])], min_count = 1,vector_size = 100, window = 5, sg=0) #CBow
model2 = gensim.models.Word2Vec([word_tokenize(sent_token[0])], min_count = 1, vector_size = 100, window = 5, sg = 1) #Skip-Gram
```

# Paragraph Embeddings : Doc2Vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_data = [TaggedDocument(word_tokenize(d), [i]) for i, d in enumerate(sent_token)]
```

```python
model = Doc2Vec(tagged_data, vector_size = 20, window = 2, min_count = 1, epochs = 100)
```

```python
model.dv.index_to_key
```

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="NLP_Seminar")
```

```python
collection.add(
    documents=["Paracetamol is used to cure fever", "I am suffering from fever"],
    metadatas=[{"source": "my_source"}, {"source": "my_source"}],
    ids=["id1", "id2"]
)
```

```python
results = collection.query(
    query_texts=["drug for fever"],
    n_results = 1
)
```
